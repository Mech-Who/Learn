```python
import torchvision

from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
```

```python
test_data = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=torchvision.transforms.ToTensor())
test_loader = DataLoader(test_data, batch_size=64, shuffle=True,  num_workers=0, drop_last=False)
```

```python
img, target = test_data[0]
img.size, target
```

```python
writer = SummaryWriter("dataloader")
for epoch in range(2):
    step = 0
    for data in test_loader:
        imgs, targets = data
        print(imgs.size(), len(targets))
        writer.add_images(f"epoch: {epoch}", imgs, step)
        step += 1
writer.close()
```
